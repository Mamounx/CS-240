```python
from __future__ import print_function, division

%matplotlib inline
from scipy.stats import norm,expon

import numpy as np
import matplotlib.pyplot as plt

import nsfg
import first
import analytic

import thinkstats2
import thinkplot
```

```python
# Q1:
resp = nsfg.ReadFemResp()
resp
```

```python
school = resp[resp["hieduc"]<=10]
university = resp[resp["hieduc"]>10]
```

```python
school_mAge = school["agemomb1"]
#The following are some statistics 
school_mean = school_mAge.mean()
school_median = school_mAge.median()
school_var = school_mAge.var()
school_std = school_mAge.std()
school_sum = school_mAge.sum()
print("school mean, median, variance, standard deviation and summation: "),school_mean, school_median, school_var, school_std, school_sum
```

```python
university_mAge = university["agemomb1"]
#The following are some statistics 
university_mean = university_mAge.mean()
university_median = university_mAge.median()
university_var = university_mAge.var()
university_std = university_mAge.std()
university_sum = university_mAge.sum()
print("school mean, median, variance, standard deviation and summation: "),university_mean, university_median, university_var, university_std, university_sum
```

```python
# Q2:
schPmf = thinkstats2.Pmf(school_mAge, label="School")
uniPmf = thinkstats2.Pmf(university_mAge, label="University")
thinkplot.PrePlot(2)
thinkplot.Hist(schPmf, align= "left", width=0.5)
thinkplot.Hist(uniPmf, align="right", width=0.5)
thinkplot.config(xlabel="Age", ylabel="PMF")
```

```python
schCdf = thinkstats2.Cdf(school_mAge, label = "School")
uniCdf = thinkstats2.Cdf(university_mAge, label = "University")
thinkplot.PrePlot(2)
thinkplot.Cdf(schCdf, color="red")
thinkplot.Cdf(uniCdf, color="blue")
thinkplot.config(xlabel="Age", ylabel="CDF")
```

```python
# Q3:
schPmf = thinkstats2.Pmf(school_mAge.diff(), label="School")
thinkplot.Hist(schPmf, width=0.5)
thinkplot.config(xlabel="Age", ylabel="PMF")
```

```python
uniPmf = thinkstats2.Pmf(university_mAge.diff(), label="University")
thinkplot.Hist(uniPmf, width=0.5)
thinkplot.config(xlabel="Age", ylabel="PMF")
```

```python
# Q4:
schCdf = thinkstats2.Cdf(school_mAge.diff(), label = "School")
thinkplot.Cdf(schCdf)
thinkplot.config(xlabel="Age", ylabel="CDF")
```

```python
uniCdf = thinkstats2.Cdf(university_mAge.diff(), label = "University")
thinkplot.Cdf(uniCdf)
thinkplot.config(xlabel="Age", ylabel="CDF")
```

```python
# Q5:
random_sample = norm.rvs(loc=0,scale=1,size=200) 
parameters = norm.fit(random_sample)
x = np.linspace(-5,5,50)
fitted_pdf = norm.pdf(resp["rostscrn"].head(50),loc = parameters[0],scale = parameters[1])
normal_dist = norm.pdf(resp["rostscrn"].head(50))
plt.plot(x,fitted_pdf,"red",label="Fitted normal dist",linestyle="dashed", linewidth=2)
plt.plot(x,normal_dist,"blue",label="Normal dist", linewidth=2)
plt.hist(random_sample,normed=1,color="cyan",alpha=.3) #alpha, from 0 (transparent) to 1 (opaque)
plt.title("Normal distribution")
plt.legend()

plt.show()
```
